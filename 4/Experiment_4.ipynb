```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
boston = pd.read_csv('HousingData.csv')
print(boston)
```

```python
boston.columns
```

```python
boston['MEDV']
```

```python
boston.fillna(0, inplace=True)   # Fills missing values in-place

x = boston.drop('MEDV', axis=1)  # Features (all columns except target)
y = boston['MEDV'] 
```

```python
x.head(10)
```

```python
y.head(10)
```

```python
reg = linear_model.LinearRegression()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
```

```python
reg.fit(x_train, y_train)
```

```python
print(reg.coef_)
```

```python
y_pred = reg.predict(x_test)
print(y_pred)
```
